# Similarity Heatmaps

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# UnifiedDiff()

```python
current_similarity = [[1, 0.74, 0.83, 1, 0.52, 0.96], 
                      [0.57, 1, 0.47, 0.71, 0.54, 0.67], 
                      [0.86, 0.64, 1, 0.86, 0.59, 0.82], 
                      [1, 0.74, 0.83, 1, 0.52, 0.96], 
                      [0.46, 0.5, 0.5, 0.46, 1, 0.42],
                      [0.81, 0.59, 0.67, 0.81, 0.41, 1]]

data = np.array(current_similarity)
labels = ['HSBC', 'AstraZeneca', 'Unilever', 'GSK', 'Tesco', 'Shell']
plt.figure(dpi=110)
plt.title('Similarity between current parsers')
sns.heatmap(data, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, 
                 yticklabels=labels, xticklabels=labels, vmin=0,vmax=1)
```
